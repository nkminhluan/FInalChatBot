# Chatbot với keras

```python
import os
urls = os.listdir(os.path.expanduser("data_chatbot"))
urls.remove(".DS_Store")
answers = []
questions = []
for url in urls:
    path = 'data_chatbot/'+url
    with open(path,encoding= 'utf-8-sig') as f:
        f_bb = f.read().splitlines()
   
    for i in range(len(f_bb)):
        a = str(f_bb[i]).split(' __eou__')
        questions.append((a[0].lower()))
        answers.append((a[1].strip().lower()))
#print(questions[1020])
print(questions[0:3])
print(answers[0:3])
```

# Tách từ dùng Underthesea

```python
from underthesea import word_tokenize
sentences = [word_tokenize(sent) for sent in questions]
print(sentences)
```

# Fit_on_texts câu

```python
from keras.preprocessing.text import Tokenizer 
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
print(tokenizer.word_index)
```

# Mode Count

```python
X_count = tokenizer.texts_to_matrix(sentences,mode='count')
print(X_count.shape)
X_count
```

# Mode Tfidf

```python
X_tfidf = tokenizer.texts_to_matrix(sentences,mode='tfidf')
print(X_tfidf.shape)
X_tfidf
```

# Chuyển đổi câu về ma trận sử dụng 2 loại: count và tfidf

```python
question_count =['thể loại phim bạn thích là gì ?'] 
sentence_count = [word_tokenize(sent) for sent in question_count]
count_vector = tokenizer.texts_to_matrix(sentence_count, mode = 'count')
tfidf_vector = tokenizer.texts_to_matrix(sentence_count, mode = 'tfidf')
print(sentence_count)
```

```python
print(tfidf_vector.shape)
print('Mode Tfidf: ',tfidf_vector)
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
def chat_count(x):
    x = [word_tokenize(i) for i in x]
    x_vector = tokenizer.texts_to_matrix(x,mode='count')
    response = '' 
    most_sim = 0  
    for i in range(len(questions)):  
        if most_sim < cosine_similarity(X_count[i].reshape(1,-1), x_vector[0].reshape(1,-1)):  
            most_sim = cosine_similarity(X_count[i].reshape(1,-1), x_vector[0].reshape(1,-1))  
            answer_index = i                    
            response = answers[answer_index] 
    return response
```

```python
def chat_tfidf(x):
    x = [word_tokenize(i) for i in x]
    x_vector = tokenizer.texts_to_matrix(x,mode='tfidf')
    response = '' 
    most_sim = 0  
    for i in range(len(questions)):  
        if most_sim < cosine_similarity(X_tfidf[i].reshape(1,-1), x_vector[0].reshape(1,-1)):  
            most_sim = cosine_similarity(X_tfidf[i].reshape(1,-1), x_vector[0].reshape(1,-1))  
            answer_index = i                    
            response = answers[answer_index] 
    return response
```

# Chatbot Sklearn

```python
from pyvi import ViTokenizer
ques = []
for i in (questions):
    ques.append(ViTokenizer.tokenize(i))
print(ques)
```

# Mode Tfidf

```python
from sklearn.feature_extraction.text import TfidfVectorizer  
vectorizer_tfidf= TfidfVectorizer()  
X_tfidf = vectorizer_tfidf.fit(ques)   
array_tfidf = X_tfidf.transform(ques).toarray()
array_tfidf
```

# Mode Hashing

```python
from sklearn.feature_extraction.text import HashingVectorizer  
vectorizer_hash= HashingVectorizer(n_features=16)  
X_hash = vectorizer_hash.fit(ques)   
array_hash = X_hash.transform(ques).toarray()
array_hash
```

```python
import numpy as np  
from sklearn.metrics.pairwise import cosine_similarity
```

```python
test_question ='hôm nay đi xem phim chứ?'
print(ViTokenizer.tokenize((test_question)))
```

```python
vector_count = X_tfidf.transform([test_question]).toarray()#.reshape(1,-1) 
print('Count: ',vector_count)
```

```python
vector_hash = X_hash.transform([test_question]).toarray()#.reshape(1,-1) 
print('Hash: ',vector_hash)
```

```python
def chat_tfidf(x):
    x = ViTokenizer.tokenize(x)
    test_vector = X_tfidf.transform([x]).toarray()
    response = '' 
    most_sim = 0  
    for i in range(len(questions)):  
        if most_sim < cosine_similarity(array_tfidf[i].reshape(1,-1), test_vector[0].reshape(1,-1)):  
            most_sim = cosine_similarity(array_tfidf[i].reshape(1,-1), test_vector[0].reshape(1,-1))  
            answer_index = i                    
            response = answers[answer_index] 
    return response
```

```python
def chat_hashing(x):
    x = ViTokenizer.tokenize(x)
    test_vector = X_hash.transform([x]).toarray()
    response = '' 
    most_sim = 0  
    for i in range(len(questions)):  
        if most_sim < cosine_similarity(array_hash[i].reshape(1,-1), test_vector[0].reshape(1,-1)):  
            most_sim = cosine_similarity(array_hash[i].reshape(1,-1), test_vector[0].reshape(1,-1))  
            answer_index = i                    
            response = answers[answer_index] 
    return response
```
