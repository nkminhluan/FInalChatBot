```python
import os
urls = os.listdir(os.path.expanduser("data_chatbot"))
urls.remove(".DS_Store")
answers = []
questions = []
for url in urls:
    path = 'data_chatbot/'+url
    with open(path,encoding= 'utf-8-sig') as f:
        f_bb = f.read().splitlines()
   
    for i in range(len(f_bb)):
        a = str(f_bb[i]).split(' __eou__')
        if len(a[0])>1:
            questions.append((a[0].lower()))
            answers.append((a[1].strip().lower()))
#print(questions[1020])
print(questions[5])
print(answers[5])
print(len(questions))
```

# Sử dụng underthesea

```python
from pyvi import ViTokenizer
from underthesea import word_tokenize
ques = []

for i in questions:
    pyvi = word_tokenize(i)
    ques.append(pyvi)
print(ques)
```

```python
import gensim
model = gensim.models.Word2Vec(sentences=ques,window=5,min_count=1,workers=4)
```

```python
def auto_chat(sms):
    sms = word_tokenize(sms.lower())
    score = []
    for i in ques:
        score.append(model.wv.n_similarity(sms,i))
        max_score = max(score)
    max_Index = score.index(max_score)
    return answers[max_Index]
```

```python
auto_chat('chào')
```

```python
#Creating GUI with tkinter
import tkinter
from tkinter import *


def send():
    msg = EntryBox.get("1.0",'end-1c').strip()
    EntryBox.delete("0.0",END)

    if msg != 'chào':
        ChatLog.config(state=NORMAL)
        ChatLog.insert(END, "You: " + msg + '\n\n')
        ChatLog.config(foreground="#442265", font=("Arial", 12 ))
        
        res = auto_chat(msg)
        ChatLog.insert(END, "Bot: " + res + '\n\n')

        ChatLog.config(state=DISABLED)
        ChatLog.yview(END)
        
    if msg == 'chào':
        ChatLog.config(state=NORMAL)
        ChatLog.insert(END, "You: " + msg + '\n\n')
        ChatLog.insert(END, 'Bot: chào bạn' +'\n\n')
        ChatLog.config(foreground="#442265", font=("Arial", 12 ))
        ChatLog.config(state=DISABLED)
        ChatLog.yview(END)

base = Tk()
base.title("Hello")
base.geometry("400x500")
base.resizable(width=FALSE, height=FALSE)

#Create Chat window
ChatLog = Text(base, bd=0, bg="light green", height="8", width="50", font="Arial",)

ChatLog.config(state=DISABLED)

#Bind scrollbar to Chat window
scrollbar = Scrollbar(base, command=ChatLog.yview, cursor="heart")
ChatLog['yscrollcommand'] = scrollbar.set

#Create Button to send message
SendButton = Button(base, font=("Verdana",12,'bold'), text="Send", width="12", height=5,
                    bd=0, activebackground='#00ff00',
                    command= send,background='white' )

#Create the box to enter message
EntryBox = Text(base, bd=0, bg="white",width="29", height="5", font="Arial")
#EntryBox.bind("<Return>", send)


#Place all components on the screen
scrollbar.place(x=376,y=6, height=386)
ChatLog.place(x=6,y=6, height=386, width=370)
EntryBox.place(x=128, y=401, height=90, width=265)
SendButton.place(x=6, y=401, height=90)

base.mainloop()
```
